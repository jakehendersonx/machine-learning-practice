```python
relevant_columns = ['attack', 'defense', 'hp', 'sp_attack', 'sp_defense', 'speed', 'base_total', 'type1', 'type2']
```

```python
data = data[relevant_columns]
```

```python
data.shape
```

```python
data.dtypes
```

```python
import matplotlib.pyplot as plt
# Loop through each column in the DataFrame and create a histogram
for column in data.columns:
    # skip categorical columns
    if column == 'type1' or column == 'type2':
        continue
    # Create a figure with a black background
    fig, ax = plt.subplots(facecolor='black')
    # plt.figure()  # Create a new figure for each histogram
    plt.hist(data[column], bins=801)  # Adjust the number of bins as needed
    plt.title(f'Histogram of {column}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    # Set the text and tick colors to white for better visibility on a black background
    plt.xticks(color='white')
    plt.yticks(color='white')
    plt.title(color='white')

    # Customize other elements like grid lines, if needed
    ax.grid(color='white', linestyle='--', linewidth=0.5)
    plt.show()
```

```python
data.to_csv('../../code/data/cleaned_pokemon.csv', index=False)
```
