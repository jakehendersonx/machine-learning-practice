```python
import nlp_helpers as hlp
from collections import Counter
```

```python
text = hlp.test_corpus_s()
cleaned = hlp.removeNonAlphaNumeric(text)
sentances = cleaned.split(".")
```

```python
counter = Counter()
for sentance in sentances:
    counter.update(sentance.split(" "))
# Iterate over the items in the Counter
for key, count in counter.items():
    print(f"{key}: {count}")

#TODO: there are non valid tokens in here at these poitns
```

```python
# length of the vocabulary
print(len(counter))
print(len(sentances))
```

```python
'''
loop through a  list two by two
get this element and next element
add next to currents count

This can be represented as a graph and adj list
each edge is a time when a word followed another
each edge will be counted and COunters know their len
'''
from collections import defaultdict, Counter
pr_graph = defaultdict(Counter)
```

```python
print(sentances)
```

```python
any_punctuation = r'[.?!]'
space_char = " "
for sentance in sentances.split(any_punctuation):
    words  = sentance.split(space_char)
    # loop through in python sliding window of 2
    for curr, next in zip(words, words[1:]):
        pr_graph[curr].update(next)
```
