```python
with open("data/english_words.txt", "r") as file:
    content = file.read()
```

```python
content
```

```python
content_ls = content.split("\n")
```

```python
content_ls
```

```python
len(content_ls)
```

```python
def reverse(word):
    l, r = 0, len(word) - 1
    flip = list(word)
    while l < r:
        flip[l], flip[r] = flip[r], flip[l]
        l += 1
        r -= 1
    return "".join(flip)
    
```

```python
reversed_content = []
for word in content_ls:
    reversed = reverse(word)
    reversed_content.append(reversed + "\n")
```

```python
reversed_content
```

```python
# write reversed content to file
with open("data/english_words_reversed.txt", "w") as file:
    file.writelines(reversed_content)
```
